```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# from bs4 import BeautifulSoup
import fastf1
# import requests
```

```python
sched=fastf1.get_event_schedule(2024)
sched["Location"]
```

```python
y,x=[],[]
for index,row in sched.iterrows():
    session = fastf1.get_session(2024, row["Location"], 'R')
    session.load()
    sesr=session.results 
    sesr["Intensity"]=sesr["GridPosition"]-sesr["Position"]
    y.append(sesr["Intensity"].max())
    print(y[-1],"----------------------------------------------------")
    x.append(row["Location"])
```

```python
npx=np.array(x)
npy=np.array(y)
for i in range(len(x)):
    print(f"{x[i]}\tintensity{y[i]}")
```

```python
plt.barh(x,y)
plt.xticks(rotation=45)
plt.xlabel("Max Intensity")
plt.ylabel("Track")
plt.title("F1 2024 Season Max Intensity on Tracks")
plt.tight_layout()
plt.savefig('max-intensity-per-track.png')
plt.show()
plt.clf()
```
